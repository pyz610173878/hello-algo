# 2.1.1 算法定义

算法是在有限时间内解决特定问题的一组指令或步骤。这里有几个限制条件需要清楚。第一有限时间，第二特定的问题，第三指令或者步骤。指令可能指得是编程语言，自然语言描述的算法用步骤更加贴切。


问题具有明确的定义和范围

# 2.1.3 数据结构与算法之间的关系

数据结构是算法的基石。数据结构为算法提供了**结构化存储的数据**，以及**操作**数据的**方法**。
结构化存储的数据比如：数组，链表，图。每种数据结构都有对应的操作数据的方法。这里的操作指得是增删改查。


算法是数据结构发挥作用的舞台。数据结构本身仅存储数据信息，结合算法才能解决特定问题。

算法通常可以基于不同的数据结构实现，但执行效率可能相差很大，选择合适的数据结构是关键。



如果脱离非结构化的数据，算法本身没有意义。


当我们在说数据结构与算法时。是在描述一个从输入处理最终输出结果的过程。
